```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.set_style('darkgrid')
sns.countplot(y='smoker', data=tips, palette='hot')
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex', data=tips)
```

```python
sns.set_style('ticks')
sns.countplot(x='time', data=tips)
sns.despine()
```

```python
plt.figure(figsize=(2.5, 4), dpi=100)

sns.set_style('ticks')
sns.countplot(x='time', data=tips).set_title('TÃ­tulo')
sns.despine(right=False, left=True, top=True, bottom=True)
```

```python
plt.figure(figsize=(2.5, 4), dpi=100)

sns.set_style('whitegrid')
sns.set_context('paper', 2)
sns.countplot(x='smoker', data=tips).set_title('TÃ­tulo')

sns.despine(right=False, left=True, top=True, bottom=False)
```
