```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.info()
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=50)
```

```python
sns.distplot(tips['tip'], kde=False, bins=3)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, height=8)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, height=8, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, height=8, kind='hist')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, height=8, kind='kde')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, height=8, kind='reg')
```

```python
sns.pairplot(tips, hue='smoker', palette='hot')
```
