```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].value_counts()
```

```python
sns.pairplot(iris, hue='species', palette='coolwarm')
```

```python
iris_pg = sns.PairGrid(iris, hue='species', palette='coolwarm')
iris_pg.map_diag(sns.histplot)
iris_pg.map_upper(sns.kdeplot)
iris_pg.map_lower(sns.scatterplot)
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips_fc = sns.FacetGrid(row='sex', col='smoker', data=tips, hue='day')
tips_fc.map(plt.scatter, 'total_bill', 'tip')
```
