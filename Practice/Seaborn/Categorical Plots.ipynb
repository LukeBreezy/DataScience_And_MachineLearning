```python
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.info()
```

```python
tips.head()
```

```python
sns.barplot(y='sex', x='total_bill', data=tips)
```

```python
import pandas as pd

dados = {
    'A': ['x'] * 3 + ['y'] * 2,
    'B': [3, 3, 3, 1, 7]
}


teste = pd.DataFrame(data=dados)
teste
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
sns.violinplot(x='total_bill', y='day', data=tips, hue='sex', split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips, hue='smoker', split=False).legend(loc=2)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, split=True)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black').legend(loc=2)
```

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='strip')
```
