```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.corr()
```

```python
sns.heatmap(tips.corr(), cmap='coolwarm', annot=True)
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
flights_pvt = flights.pivot_table(index='year', columns='month', values='passengers')
flights_pvt
```

```python
sns.heatmap(flights_pvt, linewidths=2, linecolor='lightgray', cmap='magma')
```
