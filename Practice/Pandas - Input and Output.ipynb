## Data Input and Output

```python
import pandas as pd
import numpy as np
```

```python
PS2 = pd.read_excel('files/Games.xlsx', sheet_name='PS2')
PSP = pd.read_excel('files/Games.xlsx', sheet_name='PSP')
PS3 = pd.read_excel('files/Games.xlsx', sheet_name='PS3')
PS4 = pd.read_excel('files/Games.xlsx', sheet_name='PS4')
PS5 = pd.read_excel('files/Games.xlsx', sheet_name='PS5')

Platforms = [PSP, PS5, PS3, PS4, PS2]

# Juntando os DataFrames
Games = pd.concat(Platforms, axis=0)
Games
```

```python
# Ordenando por Ano
Games.sort_values(by='Ano', inplace=True)
Games
```

```python
# Resetando o Index
Games.reset_index(inplace=True)
Games
```

```python
# Removendo a antiga coluna de Index
Games.drop('index', axis=1, inplace=True)
Games
```

```python
# Ajustando o index para iniciar no 1 e não no 0
Games.index = np.arange(1, len(Games)+1)
Games.index.names = ['Código']
Games
```

```python
# Salvando o novo DataFrame em uma nova sheet
Games.to_excel('files/GamesOutput.xlsx', sheet_name='God of War Franchise')
```

## Testes Database

```python
import yaml
import pymysql.cursors
from sqlalchemy import create_engine 
```

```python
with open('../db_credentials.yaml', 'r') as infile:
    db_crdt = yaml.safe_load(infile)

engine = create_engine(
    'mysql+pymysql://{}:{}@{}:{}/{}'.format(
        db_crdt['username'],
        db_crdt['password'],
        db_crdt['host'],
        db_crdt['port'],
        db_crdt['database']
    )
)
```

```python
Games.to_sql('GamesTable', engine)
```
