```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 5, 10)
y = x ** 2
```

## pyplot

O `pyplot` é exibe os gráficos em tela e permite customizar-los, alterando labels, cores e etc.


`plt.xlabel()`: Define o rótulo do eixo X  
`plt.ylabel()`: Define o rótulo do eixo Y


Os valores considerados nos eixos X e Y são passados por parâmetro na função `plot()`

```python
plt.xlabel('Rótulo X')
plt.ylabel('Rótulo Y')
plt.plot(x, y)
```

## subplot

Subplots são plots menores encaixados em uma mesma grade

```python
plt.subplot(1, 2, 1).plot(x, y, 'r-')
plt.subplot(1, 2, 2).plot(y, x, 'b-')
```

```python
a = np.linspace(0, 5, 15)
b = a ** 3
```

```python
plt.plot(x, y, 'r-', a, b, 'b-')
```

## plt.figure

`plt.figure()` cria uma espécie de canvas, que pode conter varios `axes` (eixos), e exibí-los

No exemplo abaixo, `fig.add_axes([])` determina as posições e tamanhos daquele eixo, através de uma lista passada por parâmetro [left, bottom, width height]  
E esses eixos podem ser plotados no canvas

```python
fig = plt.figure()

ax1 = fig.add_axes([.1, .1, 2, 2])
ax2 = fig.add_axes([.2, 1, 1, 1])


ax1.set_title('Plot Maior')
ax2.set_title('Plot Menor')

ax1.plot(x, y, 'ro-')
ax2.plot(y, x, 'b--')
```
