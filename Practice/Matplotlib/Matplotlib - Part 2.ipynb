```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 5, 10)
y = x ** 2
a = np.linspace(0, 5, 10)
b = a ** 3
```

Os subplots nada mais são que axes criados dentro da figure, só que de forma automatica, sem necessidade de posicionar ou definir tamanhos


Exemplo:

```python
# DE FORMA MANUAL
fig = plt.figure()

ax1 = fig.add_axes([.1, .1, .5, 1])
ax2 = fig.add_axes([.7, .1, .5, 1])

ax1.plot(x, y)
ax2.plot(y, x)
```

```python
# DE FORMA AUTOMATICA
fig, axes = plt.subplots(1, 2)

axes[0].set_title('Subplot 1')
axes[1].set_title('Subplot 2')


axes[0].plot(x, y)
axes[1].plot(y, x)
```

No exemplo abaixo, as variaveis `fig` e `axes` recebem respectivamente a figura (canvas que será exibido) e um array com os `axes`


## Figure Size e DPI (Dot Per Inch)

Para ajustar as dimensões da figura, podemos usar os parâmetros `figsize` e `dpi`, eles funcionam no `plt.subplots()` e no `plt.figure()`

`figsize`: Ajusta altura e largura  
`dpi`: Escalona (Aumenta ou diminui o tamanho proporcionalmente)

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 5), dpi=110)

axes[0][0].plot(x, y, 'ro-')
axes[0][1].plot(y, x, 'r--')
axes[1][0].plot(a, b, 'bo-')
axes[1][1].plot(b, a, 'b--')

plt.tight_layout()
```

Para salvar uma figura, usamos o comando `fig.savefig()` e passamos como parâmetro o caminho e nome do arquivo que será salvo (É possivel determinar o formato: png, jpg ...)

```python
fig.savefig('./files/plot_test.jpg')
```

Para incluir legendas, basta preencher o parâmetro `label` no plot

```python
fig = plt.figure(figsize=(7, 5), dpi=100)

axes = fig.add_axes([0, 0, 1, 1])
axes.plot(a, b, label='AB', color='#000000', linewidth=2, linestyle='-', marker='o', markersize=7, markerfacecolor='#FFF000', markeredgewidth=1, markeredgecolor='#000FFF')
axes.plot(x, y, label='XY', color='#444CCC', lw=2, ls='--', marker='*', ms=12, mfc='#CCC444', mew=2, mec='#AAAFFF')

# Existe uma lista de opções para esse loc, caso nenhuma das opções atenda, é possível customizar
axes.legend(loc=2)

plt.tight_layout()
```
